```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sampen import sampen2
from scipy.stats import entropy
```

```python
with h5py.File("F:\\git\\eeg_prep\\processed_files\\101_000500_000500.hdf5", "r") as f:
    A_mats = f['data']['A_mats'][:]
    A_mats_mask = f['data']['A_mask'][:]
    soz = f['metadata']["patient_info"]['soz'][:].astype(bool)
```

```python
dots = []
masked = A_mats[:,:,A_mats_mask==1]
for x in range(masked.shape[-1]):
    temp = []   
    for i in range(masked.shape[0]):
        # Normalize the vectors before calculating the dot product
        vec1 = masked[i,:,x]
        vec2 = masked[:,i,x]
        vec1_norm = vec1 / np.linalg.norm(vec1)
        vec2_norm = vec2 / np.linalg.norm(vec2)
        temp.append(np.dot(vec1_norm, vec2_norm))
    dots.append(temp)

dots = np.array(dots)
```

```python
s = soz
soz_soz = A_mats[:,:,A_mats_mask][s, :, :][:, s, :]
soz_noz = A_mats[:,:,A_mats_mask][s, :, :][:, ~s,:]
noz_soz = A_mats[:,:,A_mats_mask][~s, :, :][:, s,:]
noz_noz = A_mats[:,:,A_mats_mask][~s, :,:][:, ~s,:]
```

```python
soz_soz.shape
```

```python
def safe_entropy(x):
    # Remove any zero values to avoid log(0)
    if len(x) == 0:
        return 0  # or np.nan, depending on how you want to handle this case
    x = x + 1
    # Normalize the data to ensure it sums to 1
    p = x / np.sum(x)
    return entropy(p)
entropy_mat = np.apply_along_axis(lambda x: safe_entropy(x), axis=-1, arr=A_mats[:,:,A_mats_mask])
min_value = np.nanmin(entropy_mat[np.isfinite(entropy_mat)])
sns.heatmap(entropy_mat,vmax=entropy_mat.max(),vmin=min_value)
# remove y axis ticks and labels and put dashes where soz is
plt.yticks([])
plt.ylabel("")
soz_indices = np.where(soz)[0]
plt.yticks(soz_indices, [""] * len(soz_indices))
plt.xticks(soz_indices, [""] * len(soz_indices));
```

```python
def safe_sampen(x):
    if len(x) == 0:
        return 0  # or np.nan, depending on how you want to handle this case
    # Calculate SampEn with parameters m=2, r=0.2*std(x)
    return sampen2(x, mm=2, r=0.2*np.std(x))[1][1]
```

```python
sampen_mat = np.apply_along_axis(lambda x: safe_sampen(x), axis=-1, arr=A_mats[:,:,A_mats_mask])
```

```python
min_value = np.nanmin(sampen_mat[np.isfinite(sampen_mat)])
max_value = np.nanmax(sampen_mat[np.isfinite(sampen_mat)])

plt.figure(figsize=(7, 7))
sns.heatmap(sampen_mat, vmax=max_value, vmin=min_value, cmap='viridis')

# Remove y-axis ticks and labels and put dashes where SOZ is
plt.yticks([])
plt.ylabel("")
soz_indices = np.where(soz)[0]
plt.yticks(soz_indices, ["-"] * len(soz_indices))
plt.xticks(soz_indices, ["-"] * len(soz_indices))

plt.title("Sample Entropy Heatmap")
plt.xlabel("Channel Index")
plt.ylabel("Channel Index")

# plt.colorbar(label='Sample Entropy')
plt.tight_layout()
plt.show()
```

```python
# calculate sample entropy of each
soz_soz_sentropy = np.apply_along_axis(lambda x: sampen2(x, mm=2, r=0.2*np.std(x))[1][1], axis=-1, arr=soz_soz)
soz_noz_sentropy = np.apply_along_axis(lambda x: sampen2(x, mm=2, r=0.2*np.std(x))[1][1], axis=-1, arr=soz_noz)
noz_soz_sentropy = np.apply_along_axis(lambda x: sampen2(x, mm=2, r=0.2*np.std(x))[1][1], axis=-1, arr=noz_soz)
noz_noz_sentropy = np.apply_along_axis(lambda x: sampen2(x, mm=2, r=0.2*np.std(x))[1][1], axis=-1, arr=noz_noz)

# calculate entropy of each
soz_soz_entropy = np.apply_along_axis(lambda x: entropy(x, base=2), axis=-1, arr=soz_soz)
soz_noz_entropy = np.apply_along_axis(lambda x: entropy(x, base=2), axis=-1, arr=soz_noz)
noz_soz_entropy = np.apply_along_axis(lambda x: entropy(x, base=2), axis=-1, arr=noz_soz)
noz_noz_entropy = np.apply_along_axis(lambda x: entropy(x, base=2), axis=-1, arr=noz_noz)
```

```python
plt.figure(figsize=(20,5))
sns.heatmap(dots.T, cmap="seismic",center=0)

# Remove y-axis ticks and labels completely
plt.yticks([])
plt.ylabel("")

# Add markers for SOZ channels without labels
soz_indices = np.where(soz)[0]
plt.yticks(soz_indices, [""] * len(soz_indices))

plt.title("Dot Product of A Matrix Rows")
plt.xlabel("Time Window")

plt.show()
```

```python

# Calculate sample entropy for each column of dots
column_entropies = []
for col in dots.T:
    # Calculate sample entropy
    # Parameters: m=2 (embedding dimension), r=0.2*std (threshold)
    se = sampen2(col, mm=2, r=0.2*np.std(col))
    column_entropies.append(se[1][1])  # Use SampEn for m=2

column_entropies = np.array(column_entropies)

# Plot the sample entropies
plt.figure(figsize=(20, 5))
plt.plot(column_entropies)
plt.title("Sample Entropy of A Matrix Rows")
plt.xlabel("Channel Index")
plt.ylabel("Sample Entropy")

# Highlight SOZ channels
soz_indices = np.where(soz)[0]
plt.scatter(soz_indices, column_entropies[soz_indices], color='red', label='SOZ')

# plot median
plt.axhline(np.median(column_entropies), color='black', linestyle='--', label='Median')
# plot mean
plt.axhline(np.mean(column_entropies), color='green', linestyle='--', label='Mean')

plt.legend()
plt.tight_layout()
plt.show()

# # Print some statistics
# print(f"Mean sample entropy: {np.mean(column_entropies):.4f}")
# print(f"Median sample entropy: {np.median(column_entropies):.4f}")
# print(f"Min sample entropy: {np.min(column_entropies):.4f}")
# print(f"Max sample entropy: {np.max(column_entropies):.4f}")
# print(f"SOZ mean sample entropy: {np.mean(column_entropies[soz]):.4f}")
# print(f"Non-SOZ mean sample entropy: {np.mean(column_entropies[~soz]):.4f}")

```
